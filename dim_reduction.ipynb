```python
import numpy as np
import matplotlib.pyplot as plt

# Load dim. reduction  modules
from sklearn.decomposition import KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
```

```python
# Load dataset - Dataset of digits from 0-9
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
print(y)
```

```python
# Kernel PCA
k_pca = KernelPCA(n_components=2, kernel='linear')
X_transformed = k_pca.fit_transform(X)
# print(X_transformed.shape)

for i in range(10):
    plt.scatter(X_transformed[y==i,0],X_transformed[y==i,1])
plt.show()
```

```python
# Kernel LDA
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X, y)
X_lda = lda.transform(X)
# print(X_lda.shape)

for i in range(10):
    plt.scatter(X_lda[y==i,0],X_lda[y==i,1])
plt.show()
```

```python
# t-SNE
X_embedded = TSNE(n_components=2).fit_transform(X)
# print(X_embedded.shape)

for i in range(10):
    plt.scatter(X_embedded[y==i,0],X_embedded[y==i,1])
plt.show()
```
